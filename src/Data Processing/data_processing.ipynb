# Data Processing

## Objective
To clean and preprocess raw employee data for analysis and modeling.

## Steps Performed
- Loaded raw Excel dataset
- Performed basic inspection
- Checked missing values
- Encoded categorical variables
- Saved cleaned dataset for further analysis

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load dataset (xlsx format)
df = pd.read_excel('../../data/raw/INX_Future_Inc_Employee_Performance.xlsx')


print("Dataset loaded successfully")
print("Shape:", df.shape)

# Encode categorical variables
le = LabelEncoder()
for col in df.select_dtypes(include='object').columns:
    df[col] = le.fit_transform(df[col])

# Save processed data
df.to_csv('../../data/processed/employee_performance_processed.csv', index=False)

print("✅ Processed data saved successfully")

```
