# Performance Prediction

## Objective
To predict employee performance rating for new or existing employees using the trained model.



```python
import joblib
import pandas as pd

# Load model (correct path)
model = joblib.load('../../models/employee_performance_model.pkl')

# Load processed data
df = pd.read_csv('../../data/processed/employee_performance_processed.csv')

# Sample prediction
sample_input = df.drop('PerformanceRating', axis=1).iloc[[0]]
prediction = model.predict(sample_input)

print("Predicted Performance Rating:", prediction[0])
```
