# Model Insights and Visualization

## Objective
To visualize feature importance and understand key drivers of employee performance.


```python
import pandas as pd
import matplotlib.pyplot as plt
import joblib
import os

print("Current Directory:", os.getcwd())

# Load model (CORRECT LOCATION)
model = joblib.load('../../models/employee_performance_model.pkl')

# Load processed data
df = pd.read_csv('../../data/processed/employee_performance_processed.csv')

# Prepare features
X = df.drop('PerformanceRating', axis=1)

# Feature importance
feature_importance = pd.Series(
    model.feature_importances_,
    index=X.columns
).sort_values(ascending=False)

# Plot top 10 features
feature_importance.head(10).plot(kind='bar')

plt.title('Top Factors Affecting Employee Performance')
plt.xlabel('Features')
plt.ylabel('Importance Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
