# Results, Analysis and Business Insights

## Department-wise Performance Analysis
- The analysis indicates that R&D and Sales departments show significant variation in performance ratings.
- These departments have a higher proportion of low and high performers, indicating inconsistency.
- HR and Support departments show relatively stable performance.

## Key Factors Affecting Employee Performance
Based on feature importance and correlation analysis, the top three factors influencing employee performance are:

1. **Job Satisfaction**
   - Employees with higher job satisfaction consistently show higher performance ratings.
   - This is the most influential factor across all departments.

2. **Monthly Income**
   - Higher compensation levels are positively correlated with better performance.
   - Indicates motivation and role seniority influence performance outcomes.

3. **Work-Life Balance**
   - Employees with balanced work schedules perform better and show less variability.
   - Poor work-life balance is associated with lower performance ratings.
